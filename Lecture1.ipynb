## Finite Automata

Let's introduce with a picture

Machine has states

Formal Def: A **finite automaton** M is a 5-tuple $(Q, \Sigma,\delta,q_0,F)$

$Q = $ finite set of states

$\Sigma = $ finite set of symbols

$\delta = Q \times \Sigma \rightarrow Q$ way we will represent transitions of the machine

$q_0 \in Q = $ start state

$F \subset Q = $ set of accept states



Questions

- Is a "finite automaton" synonymous with "finite state machine"?

Say M accepts string w if

W = $w_1, w_2, ... w_n$ each $w_i \in \Sigma$ and if there is a sequence of states $r_0, r_1, ..., r_n \in Q$ where $r_0 = q_0$ for i = 0,...,n-1$.

$\delta(r_i,w_{i+1}) = r_{i+1}$

We can write down very precisely what it means for the automaton to accept its input.

L(M) = language of M = the set of $\{w \ | \ M$ accepts $w\}$

Every machine has a unique language.  There is always one language of accepted strings.  "The language that the machine recognizes."

M *recognizes* L(M).

We want to understand what kind of languages you can recognize with a finite automaton.

There is one language for a machine.  It is all the accepted strings.


An important definition...

Def: A language A is **regular** if some finite automaton M recognizes A.  (A = L(M) for some M).
