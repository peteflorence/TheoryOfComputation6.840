### P versus NP question

Can we solve searching problems without searching?

- P: 'polynomial time' -- Quickly solvable problems
- NP: 'nondeterministic polynomial time' -- quickly checkable problems.  Includes the search problems.

### The P and NP classes

- Easily checkable problems
- Easily solvable problems (is a subset of easily checkable problems)

P$=$NP or P$\ne$NP

### History of P versus NP

- 1960s: Dawn of complexity theory
 - Rabin, Blum, Hartmanis, Edmonds
 
- 1971: P versus NP and NP-completeness
 - Cook, Levin, Karp
 
- 1956: Godel writes to von Neumann
 - Remarkable letter foreshadows P versus NP
 
### Kurt Godel (1906 - 1978)

- Greatest logician of the 20th century
- Revolutionary work in the 1930s

### John von Neumann (1903 - 1957)

- Seminal work in many areas of pure and applied math,...

(They were both at the IAP at Princeton at the time!)

Godel's letter
"Suppose there was a machine which can tell whether or not there's a proof for a mathematical statement, of length n"
"The question is how fast does $\phi(n)$ grow for an optimal machine."  (This is the P vs NP question)
- He's aware of "brute force search"
- Primality... and how in general the number of steps in finite combinatorial problems can be reduced


In 50 years since Godel's letter, now thought of as holy grail problem in computer science.
