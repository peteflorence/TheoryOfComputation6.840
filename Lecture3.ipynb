We want a method for proving non-regularity which is not just: "I couldn't think of one, therefore it doesn't exist."

There is a property that all non-regular languages have.  That property is called the Pumping Lemma.





**Pumping Lemma**: If A is regular there is a p

where for $s \in A$ $|s| \ge p$ then $s = xyz$ such that

1. $xy^iz \in A \text{ for } i \ge 0$
2. $y \ne \epsilon$
3. |xy| \le p


**Proof** Let DFA M recognize A


Let $D = \{0^k 1^k \ | \ k \ge 0$

**Thm** D is not regular

**Proof**: Proof by contradiction.  Assume D is regular.  Then D has pumping length p.
